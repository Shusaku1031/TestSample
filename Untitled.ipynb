```python
im = cv2.imread("141.jpg")
for i in range(len(content["shapes"])):
    arr = np.array(content["shapes"][i]["points"])
    arr2 = []
    for a in arr:
        x = int(round(a[0]))
        y = int(round(a[1]))
        arr2.append([x,y])
    arr2 = np.array(arr2)
    
    r = random.randint(0,255)
    b = random.randint(0,255)
    g = random.randint(0,255)
    
    cv2.fillConvexPoly(im,points=arr2,color=(r,g,b))
    
plt.imshow(im)
```
